```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
```

```python
# m is the number of training examples
m = len(x_train)
print(f"number of training examples is: {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}), ({y_i})")
#String Formatting with f-strings (inserting variables into strings)
#i is a variable in x^({i}), as it can change depending on value i want ie.1,2,3
#({x_i}) is the 1.0,2.0 or y 300,500 both variables
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("housing prices")
plt.ylabel('price in 1000s of dollars')
plt.xlabel('size in sqft')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
# diff vals of w and b will give dff straight lines
```

```python
def calculate_model_output(w, b, x):
    #this is a function & it means bsically the formula for linear reg
    m = x.shape[0]
    #m is the number of total datapts. we put it to 0.
    f_wb = np.zeros(m)
    # Creates an array of zeros with length m to store the predicted values.
    #For example, if m=3, f_wb will be [0, 0, 0].
    for i in range (len(x)):
        #len(x) means function returns the number of elements in the array x.
        #For example, if x is [1000, 1500, 2000], then len(x) returns 3.
       f_wb[i] = w * x[i] + b
    return f_wb   
```

```python
tmp_f_wb = calculate_model_output(w, b, x_train)
#This stores the predicted values from the calculate_model_output function 
#using input x_train, weight w, and bias b.

plt.plot(x_train, tmp_f_wb, c='b', label='our predicition')

plt.scatter(x_train, y_train, marker='x', c='r', label='actual values')
plt.title("housing prices")
plt.ylabel('price in 1000s of dollars')
plt.xlabel('size in sqft')
plt.legend()
plt.show()
```
